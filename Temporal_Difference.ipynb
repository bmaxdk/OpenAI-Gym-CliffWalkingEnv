### Part 1: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def epsilon_greedy_probs(env, Q_s, i_episode, eps=None):
    '''
    obtains the action probabilites corresponding to epsilon-greedy policy
    CS234 Lec4 slides 18
    '''
    #set initial epsilon if None
    epsilon = 1.0/i_episode
    if eps is not None:
        epsilon = eps
    policy_s = np.ones(env.nA) * epsilon / env.nA
    idx_a_maxQ = np.argmax(Q_s)
    policy_s[idx_a_maxQ] =  1 - epsilon + (epsilon/env.nA)
    return policy_s


def update_Q(Qsa, Qsa_next, reward, alpha, gamma):
    '''
    updates the action-value function estimate using the most recent time step
    '''
    return Qsa + (alpha * (reward + (gamma*Qsa_next) - Qsa))



def sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    
    # initialize performance monitor
    plot_every = 100
    #They are also useful for tracking transactions and other pools of data where only the most recent activity is of interest.
    tmp_scores = deque(maxlen=plot_every)
    scores = deque(maxlen=num_episodes)
    
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        
        ## TODO: complete the function
        # initialize score to collect into tmp_socres
        score = 0
        
        # begin an episode, obsserve S
        state = env.reset()
        
        # get epsilon-greedy action prob
        policy_s = epsilon_greedy_probs(env, Q[state], i_episode)
        
        # pick action A from epsilon-greedy probabilities
        action = np.random.choice(np.arange(env.nA), p=policy_s)
        
        # limit number of time steps per episode
        for t_step in np.arange(300):
            
            # repeat this until S is terminal!
            # take action A and observe R,S'
            next_state, reward, done, info = env.step(action)
            # add reward to score
            score += reward
            
            if not done: # repeat this until S is terminal!
            
                # Choose action A' using policy derived from Q (e.g., epsilong-greedy)
                # get epsilon-greedy action probs
                policy_s = epsilon_greedy_probs(env, Q[next_state], i_episode)
                # pick next action A'
                next_action = np.random.choice(np.arange(env.nA), p=policy_s)
                
                # update TD estimate of Q
                Q[state][action] = update_Q(Q[state][action], Q[next_state][next_action], reward, alpha, gamma)
                
                
                # update 
                # S <-S'
                state = next_state
                # A<-A'
                action = next_action
                
            # when S is terminal;
            if done:
                # update TD estimate of Q
                Q[state][action] = update_Q(Q[state][action], 0, reward, alpha, gamma)
                #append score
                tmp_scores.append(score)
                break

                
                
        if (i_episode % plot_every == 0):
            scores.append(np.mean(tmp_scores))
            

    #plot performance
    plt.plot(np.linspace(0, num_episodes, len(scores), endpoint=False), np.asarray(scores))
    plt.xlabel('Episode Number')
    plt.ylabel('Average Reward (Over Next %d Episodes)' % plot_every)
    plt.show()

    # print best 100-episode performance
    print(('Best Average Reward over %d Episodes:' % plot_every), np.max(scores))
        
    return Q
````

````python
# obtain the estimated optimal policy and corresponding action-value function
Q_sarsa = sarsa(env, 5000, .01)

# print the estimated optimal policy
policy_sarsa = np.array([np.argmax(Q_sarsa[key]) if key in Q_sarsa else -1 for key in np.arange(48)]).reshape(4,12)
check_test.run_check('td_control_check', policy_sarsa)
print("\nEstimated Optimal Policy (UP = 0, RIGHT = 1, DOWN = 2, LEFT = 3, N/A = -1):")
print(policy_sarsa)

# plot the estimated optimal state-value function
V_sarsa = ([np.max(Q_sarsa[key]) if key in Q_sarsa else 0 for key in np.arange(48)])
plot_values(V_sarsa)
````

### Part 2: TD Control: Q-learning

In this section, you will write your own implementation of the Q-learning control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def q_learning(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    
    # Follow same process as sarsa begin  ------
    # initialize performance monitor
    plot_every = 100
    tmp_scores = deque(maxlen=plot_every)
    scores = deque(maxlen=num_episodes)
    
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        # initialize score
        score = 0
        # begin an episode, observe S
        state = env.reset()
        # Follow same process as sarsa end  ------
        
        while True:
            # Sarsamax(Q-Learning) 
            # we choose action A using policy derived from Q(e.g.,epsilon-greedy)
            # get epsilon-greed action prob
            policy_s = epsilon_greedy_probs(env, Q[state], i_episode)
            action = np.random.choice(np.arange(env.nA), p=policy_s)
            
            # then take action A and observe R and S'
            next_state, reward, done, info = env.step(action)
            # add reward to score
            score += reward
            
            # update Q using
            Q[state][action] = update_Q(Q[state][action], np.max(Q[next_state]), reward, alpha, gamma)
            
            # update S<-S'
            state = next_state
            # until S is terminal
            
            #if S is terminal
            if done:
                # append Score
                tmp_scores.append(score)
                break
        
        #follow same performance monitor steps
        if (i_episode % plot_every == 0):
            scores.append(np.mean(tmp_scores))
    #plot performance
    plt.plot(np.linspace(0, num_episodes, len(scores), endpoint=False), np.asarray(scores))
    plt.xlabel('Episode Number')
    plt.ylabel('Average Reward (Over Next %d Episodes)' % plot_every)
    plt.show()
    # print best 100-episode performance
    print(('Best Average Reward over %d Episodes: ' % plot_every), np.max(scores))
        
        
    return Q
````

### Part 3: TD Control: Expected Sarsa

In this section, you will write your own implementation of the Expected Sarsa control algorithm.

Your algorithm has four arguments:

- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.

````python
def expected_sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # Follow same process as sarsa begin  ------
    # initialize performance monitor
    plot_every = 100
    tmp_scores = deque(maxlen=plot_every)
    scores = deque(maxlen=num_episodes)
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        # initialize score
        score = 0
        # begin an episode, observe S
        state = env.reset()
        # Follow same process as q_learning end  ------
        
        # Used Expected Sarsa (Algorithm 15)
        # get epsilon-greedy action probs
        policy_s = epsilon_greedy_probs(env, Q[state], i_episode, eps=0.005)
        while True:
            # pick next action
            action = np.random.choice(np.arange(env.nA), p=policy_s)
            # take action A and observe R ans S'
            next_state, reward, done, info = env.step(action)
            # add reward to score
            score += reward
            # get epsilon-greedy action prbs for S'
            policy_s = epsilon_greedy_probs(env, Q[next_state], i_episode, eps=0.005)
            
            # update Q
            Q[state][action] = update_Q(Q[state][action], np.dot(Q[next_state], policy_s), reward, alpha, gamma)
            
            # update S <- S'
            state = next_state
            
            #until S is terminal;

            # if S is terminal
            if done:
                tmp_scores.append(score)
                break;
                
        #follow same performance monitor steps
        if (i_episode % plot_every == 0):
            scores.append(np.mean(tmp_scores))
    #plot performance
    plt.plot(np.linspace(0, num_episodes, len(scores), endpoint=False), np.asarray(scores))
    plt.xlabel('Episode Number')
    plt.ylabel('Average Reward (Over Next %d Episodes)' % plot_every)
    plt.show()
    # print best 100-episode performance
    print(('Best Average Reward over %d Episodes: ' % plot_every), np.max(scores))
        
    return Q
````
